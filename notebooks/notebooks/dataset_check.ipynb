```python
from pathlib import Path
import sys

project_root = Path.cwd().resolve()
for candidate in [project_root, *project_root.parents]:
    if (candidate / "src").exists():
        project_root = candidate
        break

if str(project_root) not in sys.path:
    sys.path.insert(0, str(project_root))

from src.miniscibert.dataset import MLMTextDataset

# Load only a sample of the corpus to test quickly
corpus_path = project_root / "data/processed/abstracts.txt"
sample_path = project_root / "data/processed/abstracts_sample.txt"
vocab_path = project_root / "models/my_tokenizer/vocab.txt"

print(f"Creating sample corpus (loading first 50MB)...")
with open(corpus_path, 'r', encoding='utf-8') as f:
    # Read up to 50MB to make it manageable
    sample_text = f.read(50 * 1024 * 1024)  # 50MB

# Write sample to temporary file
with open(sample_path, 'w', encoding='utf-8') as f:
    f.write(sample_text)

print(f"Sample corpus size: {sample_path.stat().st_size / 1024 / 1024:.2f} MB")
print(f"Loading dataset from sample...")

dataset = MLMTextDataset(
    corpus_file=str(sample_path),
    vocab_file=str(vocab_path),
    block_size=16,
    lowercase=True,
)

print("num samples:", len(dataset))
sample = dataset[0]
print(sample)
print(sample["input_ids"].shape)
print(sample["input_ids"][:10])
```
